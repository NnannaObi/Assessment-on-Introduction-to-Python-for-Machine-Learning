```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/Carol-Kelly/Hamoye-Data-Warehouse-ProjectTeam/main/earning.csv"
stats = pd.read_csv(url, delimiter=";")
```

```python
stats
```

### DATA DESCRIPTION
#### The floats signify the average hourly earnings of female and male employees.

```python
stats.isnull().sum()
```

```python
#creating a subset of the data containing female jobs and earnings
female = stats[stats.columns[0:10]]
female.rename(columns = {'femalesmanagers':'Managers','femaleprofessionals':'Professionals','femaletechniciansandtradesworkers':'Technicians and Trades workers','femalecommunityandpersonalserviceworkers':'Community and Personal Service workers','femaleclericalandadministrativeworkers':'Clerical and Administrative workers','femalesalesworkers':'Sales workers','femalemachineryoperatorsanddrivers':'Machinery Operators and Drivers','femalelabourers':'Labourers'},inplace=True)
```

```python
list_df = []
for i in female["year"]:
    x = female[female["year"] == i].T.reset_index()
    x["year"] = i
    x["gender"] = 'Female'
    x.drop(index=x.index[0:2],axis=0,inplace=True)
    list_df.append(x)
for df in list_df: #changing the column names for each dataframe in the list
    df.columns = ["Job_title","earnings","year","gender"]
f = pd.concat(list_df)
f.reset_index(drop=True,inplace=True)
f
```

```python
#creating a subset of the data containing male jobs and earnings
male = stats.iloc[:,np.r_[0,10:19]]
male.rename(columns = {'malemanagers':'Managers','maleprofessionals':'Professionals','maletechniciansandtradesworkers':'Technicians and Trades workers','malecommunityandpersonalserviceworkers':'Community and Personal Service workers','maleclericalandadministrativeworkers':'Clerical and Administrative workers','malesalesworkers':'Sales workers','malemachineryoperatorsanddrivers':'Machinery Operators and Drivers','malelabourers':'Labourers'},inplace=True)
male
```

```python
list_df = []
for i in male["year"]:
    x = male[male["year"] == i].T.reset_index()
    x["year"] = i
    x["gender"] = 'Male'
    x.drop(index=x.index[0:2],axis=0,inplace=True)
    list_df.append(x)
for df in list_df: #changing the column names for each dataframe in the list
    df.columns = ["Job_title","earnings","year","gender"]
m = pd.concat(list_df)
m.reset_index(drop=True,inplace=True)
m
```

```python
#creating a subset of the data containing both sexes' jobs and earnings
persons = stats.iloc[:,np.r_[0,19:28]]
persons.rename(columns = {'personmanagers':'Managers','personprofessionals':'Professionals','persontechniciansandtradesworkers':'Technicians and Trades workers','personcommunityandpersonalserviceworkers':'Community and Personal Service workers','personclericalandadministrativeworkers':'Clerical and Administrative workers','personsalesworkers':'Sales workers','personmachineryoperatorsanddrivers':'Machinery Operators and Drivers','personlabourers':'Labourers'},inplace=True)
persons
```

```python
list_df = []
for i in persons["year"]:
    x = persons[persons["year"] == i].T.reset_index()
    x["year"] = i
    x["gender"] = 'Persons'
    x.drop(index=x.index[0:2],axis=0,inplace=True)
    list_df.append(x)
for df in list_df: #changing the column names for each dataframe in the list
    df.columns = ["Job_title","earnings","year","gender"]
p = pd.concat(list_df)
p.reset_index(drop=True,inplace=True)
p
```

```python
#concatenating the three datasets- female, male and both sexes- together to form one dataset
data = pd.concat([f,m,p]).reset_index(drop=True)
data.sort_values(by='year')
```
